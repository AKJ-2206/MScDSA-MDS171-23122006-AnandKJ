```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("data_clean.csv")
df = df.drop("Timestamp" , axis = 1)
df = df.drop("Score" , axis = 1)
df = df.drop("Email Address" , axis = 1)
df = df.drop("Name of the Student" , axis = 1)
df = df.rename(columns = {"Do you own a private vehicle?" : "P_vehicle" , "What type of transportation do you prefer when coming to college?" : "Transportation" , "Are you willing to buy a private vehicle in future?" : "buy_pvehicle" , "How much money do you spend for petrol for a month? (If you have private vehicle?)" : "petrol_expense" , "What type of vehicle do you own? (If you have private vehicle)" : "pvehicle_type" , "How much distance do you travel in a monthly bases?" : "travel_expense" , "How much are you satisfied with the collage bus service?" : "Satisfactions" , "Where do you stay?" : "Stay" , "By what time do you avail the bus service in the morning mostly?" : "bus_time" , "How much time will it take you to reach till your residence/villa from collage?" : "reachtime"})
df = df.to_csv("data_columns.csv" , index = False)
```

```python
df = pd.read_csv("data_columns.csv")

# Travel Expense
df["travel_expense"] = df["travel_expense"].replace("<200 km" , 200)
df["travel_expense"] = df["travel_expense"].replace("<200" , 200)
df["travel_expense"] = df["travel_expense"].replace("<500 km" , 500)
df["travel_expense"] = df["travel_expense"].replace("<1000 km" , 1000)
df["travel_expense"] = df["travel_expense"].replace("<<2000 km" , 2000)

# Petrol expense
df["petrol_expense"] = df["petrol_expense"].replace("<1000" , 1000)
df["petrol_expense"] = df["petrol_expense"].replace("<500" , 500)
df["petrol_expense"] = df["petrol_expense"].replace("<<2000" , 2000)
df["petrol_expense"] = df["petrol_expense"].replace("<2000" , 2000)
df["petrol_expense"] = df["petrol_expense"].replace("<5000" , 5000)

# Satisfaction
df["Satisfactions"] = df["Satisfactions"].replace("1 (Not good)" , 1)
df["Satisfactions"] = df["Satisfactions"].replace("2 (Could be better)" , 2)
df["Satisfactions"] = df["Satisfactions"].replace("3 (Average)" , 3)
df["Satisfactions"] = df["Satisfactions"].replace("4 (Good)" , 4)
df["Satisfactions"] = df["Satisfactions"].replace("5 (best)" , 5)

#bus_time
df["bus_time"] = df["bus_time"].replace("7:30 am" , 7.30)
df["bus_time"] = df["bus_time"].replace("7:50 am" , 7.50)
df["bus_time"] = df["bus_time"].replace("8:05 am" , 8.05)
df["bus_time"] = df["bus_time"].replace("8:20 am" , 8.20)
df["bus_time"] = df["bus_time"].replace("8:40 am" , 8.40)

#Duration
df["reachtime"] = df["reachtime"].replace("2 minutes" , 2)
df["reachtime"] = df["reachtime"].replace("5 minutes" , 5)
df["reachtime"] = df["reachtime"].replace("10 minutes" , 10)
df["reachtime"] = df["reachtime"].replace("15 minutes" , 15)
df["reachtime"] = df["reachtime"].replace("20 minutes" , 20)
df["reachtime"] = df["reachtime"].replace("25 minutes" , 25)
df["reachtime"] = df["reachtime"].replace("25. minutes" , 25)
df["reachtime"] = df["reachtime"].replace("30 minutes" , 30)

#Course
df["Course "] = df["Course "].replace("Master of Science (MSc) in Data Science + Master of Science (MS) in System Science" , "MSc.DS")
df["Course "] = df["Course "].replace("MSc (Data Science)" , "MSc.DS")
df["Course "] = df["Course "].replace("BBA (Business Analytics)" , "BBA")
df["Course "] = df["Course "].replace("LLM (Corporate & Commercial Law)" , "LLM")
df["Course "] = df["Course "].replace("MBA" , "MBA")
df["Course "] = df["Course "].replace("BBA (Business Analytics/Honours/Honours with Research)" , "BBA")
df["Course "] = df["Course "].replace("BCom (Financial Analytics/Honours/Honours with Research)" , "BCom")
df["Course "] = df["Course "].replace("MBA" , "MBA")
df["Course "] = df["Course "].replace("MSc (Economics & Analytics)" , "MSc.EA")
df["Course "] = df["Course "].replace("BBA LLB (Honours)" , "BBA.LLB")
df["Course "] = df["Course "].replace("BSc (Economics & Analytics)" , "BSc.EA")
df["Course "] = df["Course "].replace("BA LLB (Honours)" , "BA.LLB")
df["Course "] = df["Course "].replace("BSc (Data Science)" , "BSc.DS")
df["Course "] = df["Course "].replace("MBA" , "MBA")
df["Course "] = df["Course "].replace("MBA" , "MBA")
df["Course "] = df["Course "].replace("MBA" , "MBA")
df["Course "] = df["Course "].replace("MBA" , "MBA")

# Vehicle Type
df["pvehicle_type"] = df["pvehicle_type"].replace("2 Wheeler" , 2)
df["pvehicle_type"] = df["pvehicle_type"].replace("4 Wheeler" , 4)


df = df.to_csv("data_clean.csv" , index = False)

```

```python
male_data = df[df['Gender'] == 'Male']
female_data = df[df['Gender'] == 'Female']
male_satisfaction = male_data['Course ']
female_satisfaction = female_data['Course ']

plt.figure(figsize=(10, 4))
plt.hist(male_satisfaction, bins=30, alpha=1, label='Male' , align="left")
plt.hist(female_satisfaction, bins=30, alpha=1, label='Female',align="right")
plt.xlabel('Satisfaction Level')
plt.ylabel('Frequency')
plt.title('Satisfaction Level Histogram by Gender')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='Gender', hue='Transportation', data=df)
plt.xlabel('Gender')
plt.ylabel('Mode of Transportation')
plt.title('Gender-wise differenciation in mode of transportation')
plt.xticks(rotation=45)
plt.legend(title='Platform')
plt.show()
```
